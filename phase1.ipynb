```python
import networkx as nx
import random
import matplotlib.pyplot as plt
from logbin import logbin
```

```python
G = nx.Graph()

n = 10000

G.add_node(1)

for i in range(n-1):
    new_node = max(G.nodes) + 1
    if len(G.nodes) == 1:
        target_node = 1
    else:
        target_node = random.choice(random.choice(tuple(G.edges)))
    G.add_node(new_node)
    G.add_edge(new_node, target_node)

# nx.draw(G)
# nx.draw_networkx(G)
```

```python
degrees = [len(G.adj[i]) for i in G.adj]

binned_degrees = logbin(degrees, scale=1.3)

plt.plot(binned_degrees[0], binned_degrees[1])
plt.yscale('log')
plt.xscale('log')
plt.show()
```
